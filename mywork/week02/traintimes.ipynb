```python
from xml.dom.minidom import parseString
import requests
import csv

# Define the tags you want to retrieve from the XML
retrieveTags = [
    'TrainStatus',
    'TrainLatitude',
    'TrainLongitude',
    'TrainCode',
    'TrainDate',
    'PublicMessage',
    'Direction'
]

# URL for the Irish Rail API
url = "http://api.irishrail.ie/realtime/realtime.asmx/getCurrentTrainsXML"
page = requests.get(url)

# Parse the XML content using minidom
doc = parseString(page.content)

# Check if it works, output the formatted XML to the console
# This step is optional; you can remove it once you confirm the data is correct
print(doc.toprettyxml(newl=''))

# Save the raw XML to a file (optional)
with open("trainxml.xml", "w") as xmlfp:
    doc.writexml(xmlfp)

# Open a CSV file to store the filtered data
with open('traintimes_D.csv', mode='w', newline='') as train_file:
    train_writer = csv.writer(train_file, delimiter='\t', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    # Write the header row to the CSV
    train_writer.writerow(retrieveTags)

    # Get the train positions from the XML
    objTrainPositionsNodes = doc.getElementsByTagName("objTrainPositions")
    
    # Loop through each train position in the XML
    for objTrainPositionsNode in objTrainPositionsNodes:
        # Extract the TrainCode (this is what we will filter on)
        traincodenode = objTrainPositionsNode.getElementsByTagName("TrainCode").item(0)
        traincode = traincodenode.firstChild.nodeValue.strip()
        
        # Only process trains whose TrainCode starts with "D"
        if traincode.startswith("D"):
            dataList = []
            
            # Retrieve all the tags and their values for this train
            for retrieveTag in retrieveTags:
                datanode = objTrainPositionsNode.getElementsByTagName(retrieveTag).item(0)
                dataList.append(datanode.firstChild.nodeValue.strip())

            # Write the filtered train data to the CSV file
            train_writer.writerow(dataList)
```
